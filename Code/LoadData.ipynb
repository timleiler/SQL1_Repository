```python
import pandas as pd
import sqlite3
import sys
from pathlib import Path
```

```python
#Point to project root so config.py is findable
sys.path.append(str(Path().resolve().parent))
from config import DATA_FILE, DB_FILE
```

```python
#First I read the excel sheet into pandas
#I create empty dictionary first
dfs = {}  

dfs["gl"]          = pd.read_excel(DATA_FILE, sheet_name="GL")
dfs["coa"]         = pd.read_excel(DATA_FILE, sheet_name="COA")
dfs["calendar"]    = pd.read_excel(DATA_FILE, sheet_name="Calendar")
dfs["territory"]   = pd.read_excel(DATA_FILE, sheet_name="Territory")
dfs["cashflow"]    = pd.read_excel(DATA_FILE, sheet_name="CashFlow_St")
dfs["soce_st"]     = pd.read_excel(DATA_FILE, sheet_name="SoCE_St")


for name, df in dfs.items():
    print(f"{name}: {type(df)} — {df.shape}")

```

```python
#Now I create a database file

conn = sqlite3.connect(DB_FILE)

#I write the dataframe into the database as tables
for sheets, dfs in dfs.items():
    dfs.to_sql(sheets.lower(), conn, if_exists="replace", index=False)
    

conn.close()
```
